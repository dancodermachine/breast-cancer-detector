# Project: Breast Cancer Detector

This project aims to use different algorithms to detect breast cancer and compare their accuracy scores and runtimes to find the most efficient algorithm. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import scale

from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import (VotingClassifier, BaggingClassifier, RandomForestClassifier,
AdaBoostClassifier, GradientBoostingClassifier)

from xgboost import XGBClassifier

from sklearn.metrics import classification_report, accuracy_score

import time
```

## 1. Data Collection

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"

head_names = ['id', 'clump_thickness', 'uniform_cell_size', 'uniform_cell_shape', 'marginal_adhesion',
              'single_epithelial_size', 'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class']

dataframe = pd.read_csv(url, names=head_names)
```

## 2. Data Cleaning

```python
data = dataframe

data.head()
```

```python
data.info()
```

```python
data['bare_nuclei'].replace('?', -999999, inplace=True)
data['bare_nuclei'] = data['bare_nuclei'].astype('int')
```

```python
data.info()
```

```python
data.describe()
```

```python
data.corr()
```

```python
sns.pairplot(data);
```

## 3. Machine Learning

```python
seed = 101096

dataset = data

X = dataset.drop(['id', 'class'], axis=1).values
y = data['class'].values

X_scaled = scale(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,
                                                   random_state=seed, stratify=y)
```

### 3.1 K Neighors Classifier 

```python
knn_start = time.time()
knn = KNeighborsClassifier()

knn_parameters = {'n_neighbors': np.arange(1,20),
                  'weights': ['uniform', 'distance'],
                  'leaf_size': np.arange(20,40)}

knn_cv = GridSearchCV(knn, param_grid = knn_parameters, cv = 5, n_jobs=-1)

knn_cv.fit(X_train, y_train)

knn_y_pred = knn_cv.predict(X_test)

knn_accuracy_score = accuracy_score(y_test, knn_y_pred) 

print("Accuracy score: " + str(knn_accuracy_score))
print("Best parameters: " + str(knn_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, knn_y_pred))
knn_end = time.time()
knn_time = knn_end-knn_start
print("Time to run algorithm: " + str(knn_time))
```

### 3.2 Support Vector Machines

```python
svm_start = time.time()
svm = SVC(random_state=seed, kernel='linear')

svm_parameters = {'gamma': [0.1, 1, 10, 100],
                  'C': [0.1, 1, 10, 100, 1000],
                  'degree': np.arange(1,6,1)}

svm_cv = GridSearchCV(svm, param_grid = svm_parameters, cv = 5, n_jobs=-1) 
svm_cv.fit(X_train, y_train)

svm_y_pred = svm_cv.predict(X_test)

svm_accuracy_score = accuracy_score(y_test, svm_y_pred) 

print("Accuracy score: " + str(svm_accuracy_score))
print("Best parameters: " + str(svm_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, svm_y_pred))
svm_end = time.time()
svm_time = svm_end-svm_start
print("Time to run algorithm: " + str(svm_time))
```

### 3.3 Decision Trees

```python
dt_start = time.time()
dt = DecisionTreeClassifier(random_state=seed)

dt_parameters = {'criterion':['gini','entropy'],
                 'max_depth':np.arange(1,10),
                 'min_samples_split': np.arange(1,40,1),
                 'max_features':np.arange(0.1, 0.8,0.1),
                 'min_samples_leaf': np.arange(1,5,1)}

dt_cv = GridSearchCV(dt, param_grid=dt_parameters, cv=5, n_jobs=-1) 

dt_cv.fit(X_train, y_train)

dt_y_pred = dt_cv.predict(X_test)

dt_accuracy_score = accuracy_score(y_test, dt_y_pred) 

print("Accuracy score: " + str(dt_accuracy_score))
print("Best parameters: " + str(dt_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, dt_y_pred))
dt_end = time.time()
dt_time = dt_end-dt_start
print("Time to run algorithm: " + str(dt_time))
```

### 3.4 Voting Classifier

```python
vc_start = time.time()
knn_el = KNeighborsClassifier(leaf_size = 20, n_neighbors = 12, weights = 'distance')
svm_el = SVC(random_state=seed, kernel='linear', gamma=0.1, degree=1, C=0.1)
dt_el = DecisionTreeClassifier(criterion='gini', max_depth=8, random_state=seed, max_features=0.7,
                               min_samples_leaf=1, min_samples_split=2)

classifiers_vc = [('Logistic Regression', svm_el),
               ('K Nearest Neighours', knn_el),
               ('Classification Tree', dt_el)]

vc = VotingClassifier(estimators=classifiers_vc)

vc.fit(X_train, y_train)

vc_y_pred = vc.predict(X_test)

vc_accuracy_score = accuracy_score(y_test, vc_y_pred) 

print("Accuracy score: " + str(vc_accuracy_score))
print('---------------------------------------------')
print(classification_report(y_test, vc_y_pred))
vc_end = time.time()
vc_time = vc_end-vc_start
print("Time to run algorithm: " + str(dt_time))
```

**Note:** It is necessary to add the runtime of the *KNN*, *SVM* and *Decision Tree* algorithms to the runtime of this cell since this algorithm is using the best parameters found the in the previous cells.

### 3.5 Bagging Classifier

```python
bc_start = time.time()

bc_param = {'n_estimators': np.arange(100, 400, 5)}


dt_2 = DecisionTreeClassifier(criterion='gini', max_depth=8, random_state=seed, max_features=0.7,
                               min_samples_leaf=1, min_samples_split=2)
bc = BaggingClassifier(base_estimator = dt_2)

bc_cv = GridSearchCV(bc, param_grid=bc_param, cv=5, n_jobs=-1)

bc_cv.fit(X_train, y_train)
bc_y_pred = bc_cv.predict(X_test)

bc_accuracy_score = accuracy_score(y_test, bc_y_pred) 

print("Accuracy score: " + str(bc_accuracy_score))
print("Best parameters: " + str(bc_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, bc_y_pred))
bc_end = time.time()
bc_time = bc_end-bc_start
print("Time to run algorithm: " + str(bc_time))
```

### 3.6 Random Forest

```python
rf_start = time.time()
rf = RandomForestClassifier(random_state=seed, criterion='gini', max_depth=8, max_features=0.7,
                            min_samples_leaf=1, min_samples_split=2)


rf_params = {'n_estimators' : np.arange(50,400,50)}
          
          
rf_cv = GridSearchCV(rf, param_grid = rf_params, cv = 5, n_jobs=-1)

rf_cv.fit(X_train, y_train)
rf_y_pred = rf_cv.predict(X_test)

rf_accuracy_score = accuracy_score(y_test, rf_y_pred) 

print("Accuracy score: " + str(rf_accuracy_score))
print("Best parameters: " + str(rf_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, rf_y_pred))
rf_end = time.time()
rf_time = rf_end-rf_start
print("Time to run algorithm: " + str(rf_time))
```

### 3.7 AdaBoost

```python
adab_start = time.time()

dt_3 = DecisionTreeClassifier(max_depth=3, random_state=seed)

adab = AdaBoostClassifier(base_estimator=dt_3)

adab_params = {'n_estimators' : np.arange(50,400,50)}
                
adab_cv = GridSearchCV(adab, param_grid = adab_params, cv = 5, n_jobs=-1)

adab_cv.fit(X_train, y_train)
adab_y_pred = adab_cv.predict(X_test)

adab_accuracy_score = accuracy_score(y_test, adab_y_pred) 

print("Accuracy score: " + str(adab_accuracy_score))
print("Best parameters: " + str(adab_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, adab_y_pred))
adab_end = time.time()
adab_time = adab_end-adab_start
print("Time to run algorithm: " + str(adab_time))
```

### 3.8 Stochastic Gradient Boosting 

```python
sgbt_start = time.time()

sgbt = GradientBoostingClassifier(random_state = seed, subsample=0.8)

sgbt_params = {'max_depth': np.arange(1,10,1),
              'max_features': ['log2', 'sqrt'],
              'min_samples_leaf': np.arange(0.1, 0.5, 0.05),
              'n_estimators': np.arange(50,400,50)}
                
sgbt_cv = GridSearchCV(sgbt, param_grid = sgbt_params, cv = 5, n_jobs=-1)

sgbt_cv.fit(X_train, y_train)
sgbt_y_pred = sgbt_cv.predict(X_test)

sgbt_accuracy_score = accuracy_score(y_test, sgbt_y_pred) 

print("Accuracy score: " + str(sgbt_accuracy_score))
print("Best parameters: " + str(sgbt_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, sgbt_y_pred))
sgbt_end = time.time()
sgbt_time = sgbt_end-sgbt_start
print("Time to run algorithm: " + str(sgbt_time))
```

### 3.9 XGBoost

```python
xgb_start = time.time()

xgb = XGBClassifier(random_state = seed, objective='binary:logistic')

xgb_params = {'n_estimators': np.arange(10,100,10),
              'learning_rate': np.arange(0.05,1.05,.05),
              'subsample': [0.3, 0.4, 0.5, 0.9],
              'max_depth':np.arange(1,10)}
                
xgb_cv = GridSearchCV(xgb, param_grid = xgb_params, cv = 5, n_jobs=-1)

xgb_cv.fit(X_train, y_train)
xgb_y_pred = xgb_cv.predict(X_test)

xgb_accuracy_score = accuracy_score(y_test, xgb_y_pred) 

print("Accuracy score: " + str(xgb_accuracy_score))
print("Best parameters: " + str(xgb_cv.best_params_))
print('---------------------------------------------')
print(classification_report(y_test, xgb_y_pred))
xgb_end = time.time()
xgb_time = xgb_end-xgb_start
print("Time to run algorithm: " + str(xgb_time))
```

## 4. Summary

```python
summary_dic = {'Algorithm':['KNN', 'SVM', 'Decision Tree', 'Voting Classifier', 'Bagging Classifier',
                            'Random Forest', 'Stochastic Gradient Boosting', 'XGBoost'],
               'Accuracy Score':[knn_accuracy_score, svm_accuracy_score, dt_accuracy_score, vc_accuracy_score, bc_accuracy_score,
                        rf_accuracy_score, sgbt_accuracy_score, xgb_accuracy_score],
               'Runtime':[knn_time, svm_time, dt_time, vc_time + knn_time + svm_time + dt_time, bc_time, rf_time, sgbt_time, xgb_time]}

dataframe_summary = pd.DataFrame(summary_dic).sort_values(by='Accuracy Score', ascending=False).reset_index(drop=True)

dataframe_summary
```

**Conclusion:** The top 3 algorithms were *Voting Classifier*, *SVM*, and *Stochastic Gradient Boosting* with accuracy scores of 0.967, 0.962 and 0.962 respectively. However, among those 3 algorithms, the fastest was SVM. Therefore it can be concluded that the most efficient algorithm was SVM.

**Note:** Ensembly methods such as XGBoost or Random Forest (and others) are known in industry to outperform algorithms such as SVM, KNN and Decision Trees. However, during this project, not many values were used during hyperparameter tuning since it can take long time. Therefore, there is a high probability that ensemble methods used in this project can outperform the SVM if we add more hyperparameters in the GridSearch. Further investigation needs to be done.
